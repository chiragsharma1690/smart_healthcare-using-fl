```python
#### Extracting MFCC's For every audio file
import pandas as pd
import os
import librosa

root_dir = "C:/Users/chira/Downloads/Federated Learning Project/my_code_exp/ADReSSo21"

ad_audio_dataset_path = root_dir + "/diagnosis/train/audio/ad_segmented"
cn_audio_dataset_path = root_dir + "/diagnosis/train/audio/cn_segmented"
train_audio_dataset_path = root_dir + "/diagnosis/train/audio"
test_audio_dataset_path = root_dir + "/diagnosis/test-dist"
train_mmse = root_dir + "/diagnosis/train/adresso-train-mmse-scores.csv"
```

```python
# feature extraction for one audio file
def features_extractor(file_name):

    audio, sample_rate = librosa.load(file_name, res_type='kaiser_fast')
    mfccs_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
    mfccs_scaled_features = np.mean(mfccs_features.T,axis=0)
    
    return mfccs_scaled_features
```

```python
# feature extraction for whole train files
import pandas as pd
from tqdm import tqdm
import os
import numpy as np

df = pd.read_csv(train_mmse)

extracted_features = []
for i, row in tqdm(df.iterrows()):
    
    file_name = os.path.join(os.path.abspath(train_audio_dataset_path), str(row["dx"]) + '_segmented/', str(row["adressfname"]) + '.wav')
    mmse_scores = row["mmse"]
    class_folder = row["dx"]
    if class_folder == 'ad':
        class_value = 1
    else:
        class_value = 0

    data = features_extractor(file_name)

    # print(wav)

    extracted_features.append([data, class_value, mmse_scores])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
from tensorflow.keras.optimizers import Adam
from sklearn import metrics
```

```python
### No of classes
# num_labels = 2
```

```python
model=Sequential()
###first layer
model.add(Dense(100,input_shape=(40,)))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###second layer
model.add(Dense(200))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###third layer
model.add(Dense(100))
model.add(Activation('relu'))
model.add(Dropout(0.5))

###final layer
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
## Trianing my model
from tensorflow.keras.callbacks import ModelCheckpoint
from datetime import datetime 

num_epochs = 10000
num_batch_size = 20

checkpointer = ModelCheckpoint(filepath='saved_models/audio_classification.hdf5', 
                               verbose=1, save_best_only=True)
print("running1")
start = datetime.now()

model.fit(X_train, y_train, batch_size = num_batch_size, epochs = num_epochs, validation_data= (X_test, y_test) , callbacks = [checkpointer], verbose=1)


duration = datetime.now() - start
print("Training completed in time: ", duration)
```

```python
test_accuracy=model.evaluate(X_test,y_test,verbose=0)
print(test_accuracy[1])
```
